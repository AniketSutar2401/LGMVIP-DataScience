# Task 4: Image to Pencil Sketch with Python

# Importing Libraries

```python
from google.colab.patches import cv2_imshow
import cv2
```

# Loading the image

```python
img = cv2.imread('IMG.jpg', 1)
#img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv2_imshow(img)
```

# Converting an image into gray_scale image

```python
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img_grey)
```

# Inverting the image

```python
img_invert = cv2.bitwise_not(img_grey)
cv2_imshow(img_invert)
```

## Smoothing the image 

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),0)

cv2_imshow(img_smoothing)
```

# Inverting the blured image

```python
invertedblur = cv2.bitwise_not(img_smoothing)
cv2_imshow(invertedblur)
```

# Pencil Sketch


```python
sketch = cv2.divide(img_grey, invertedblur, scale=256.0)

cv2_imshow(sketch)
```

# Final Output Original and Pencil Sketch

```python
cv2_imshow(img)
cv2_imshow(sketch)
```
